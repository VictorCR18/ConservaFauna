```python
# Importar a extensão do IPython SQL Magic
%load_ext sql

# Conectar ao banco de dados
%sql postgresql://postgres:admin@localhost/ConservaFauna

# Criar a janela principal do Tkinter
window = tk.Tk()

# Centralizar a janela principal
window.eval('tk::PlaceWindow . center')

# Criar a janela para exibir os registros
window_registros = tk.Toplevel(window)

# Variável para armazenar a escolha do usuário
escolha = None

# Funções de consulta
def consultar_especie():
    global escolha
    escolha = "Especie"
    messagebox.showinfo("Consulta", "Consultar por Espécie")
    exibir_registros()

def consultar_especies_ameacadas():
    global escolha
    escolha = "especies_ameacadas"
    messagebox.showinfo("Consulta", "Consultar por Espécie em perigo")
    exibir_registros()

def consultar_especie_conserva():
    global escolha
    escolha = "especie_conserva"
    messagebox.showinfo("Consulta", "Consultar por conserva")
    exibir_registros()

def consultar_ameacas():
    global escolha
    escolha = "Ameacas"
    messagebox.showinfo("Consulta", "Consultar por Ameaças")
    exibir_registros()

def consultar_localizacao():
    global escolha
    escolha = "Localizacao"
    messagebox.showinfo("Consulta", "Consultar por Localização")
    exibir_registros()

def consultar_avistamento():
    global escolha
    escolha = "Avistamentos"
    messagebox.showinfo("Consulta", "Consultar por Registro")
    exibir_registros()

def consultar_avistamento_recente():
    global escolha
    escolha = "avistamentos_recentes"
    messagebox.showinfo("Consulta", "Consultar por Avistamento recente")
    exibir_registros()

def consultar_instituicao():
    global escolha
    escolha = "Instituicao"
    messagebox.showinfo("Consulta", "Consultar por Instituição")
    exibir_registros()

def consultar_especies_amazonia():
    global escolha
    escolha = "especies_amazonia"
    messagebox.showinfo("Consulta", "Consultar por Espécies da região amazonica")
    exibir_registros()
    

# Função para exibir os registros em uma tabela
def exibir_registros():
    try:
        # Executar a consulta usando a mágica %sql
        result = %sql SELECT * FROM {escolha}

        # Obter o resultado como um DataFrame do Pandas
        df = result.DataFrame()

        # Limpar a janela de registros antes de exibir os novos registros
        for widget in window_registros.winfo_children():
            widget.destroy()

        # Criar o Treeview na janela de registros
        treeview = ttk.Treeview(window_registros)
        treeview["columns"] = list(df.columns)
        treeview["show"] = "headings"

        # Definir as colunas
        for column in df.columns:
            treeview.heading(column, text=column)

        # Adicionar os registros
        for row in df.itertuples(index=False):
            treeview.insert("", "end", values=row)

        # Exibir o Treeview
        treeview.pack()

        # Criar botão "Voltar"
        btn_voltar = tk.Button(window_registros, text="Voltar", command=voltar)
        btn_voltar.pack()

        # Exibir a janela de registros
        window_registros.deiconify()

    except Exception as error:
        messagebox.showerror("Erro", f"Erro ao executar a consulta: {error}")

# Função para voltar à janela principal
def voltar():
    window_registros.withdraw()

# Esconder a janela de registros inicialmente
window_registros.withdraw()

# Criar a barra de menu
menu_bar = tk.Menu(window)


#######################################################################################################################################


def inserir_especie():
    global escolha
    escolha= "Especie"
    # Criar a janela de inserção
    window_inserir_especie = tk.Toplevel(window)
    
    # Criar os campos de entrada
    lbl_nome_cientifico = tk.Label(window_inserir_especie, text="Nome cientifico:")
    entry_nome_cientifico = tk.Entry(window_inserir_especie)

    lbl_nome_popular = tk.Label(window_inserir_especie, text="Nome popular:")
    entry_nome_popular = tk.Entry(window_inserir_especie)

    lbl_descricao = tk.Label(window_inserir_especie, text="Descrição:")
    entry_descricao = tk.Entry(window_inserir_especie)
    
    lbl_habitat = tk.Label(window_inserir_especie, text="Habitat:")
    entry_habitat = tk.Entry(window_inserir_especie)

    lbl_classificacao = tk.Label(window_inserir_especie, text="Classificacao de risco:")
    entry_classificacao = tk.Entry(window_inserir_especie)
    
    lbl_id_ame = tk.Label(window_inserir_especie, text="ID da ameaça:")
    entry_id_ame = tk.Entry(window_inserir_especie)
    
    # Posicionar os campos de entrada
    lbl_nome_cientifico.grid(row=0, column=0)
    entry_nome_cientifico.grid(row=0, column=1)
    lbl_nome_popular.grid(row=1, column=0)
    entry_nome_popular.grid(row=1, column=1)
    lbl_descricao.grid(row=2, column=0)
    entry_descricao.grid(row=2, column=1)
    lbl_habitat.grid(row=3, column=0)
    entry_habitat.grid(row=3, column=1)
    lbl_classificacao.grid(row=4, column=0)
    entry_classificacao.grid(row=4, column=1)
    lbl_id_ame.grid(row=5, column=0)
    entry_id_ame.grid(row=5, column=1)

    
    # Função para salvar o novo registro
    def salvar_registro():
        try:
            # Obter os valores dos campos de entrada
            nome_cientifico_value = entry_nome_cientifico.get()
            nome_popular_value = entry_nome_popular.get()
            descricao_value = entry_descricao.get()
            habitat_value = entry_habitat.get()
            classificacao_value = entry_classificacao.get()
            id_ame_value = entry_id_ame.get()
            
            # Conectar ao banco de dados
            conn = pg.connect("postgresql://postgres:admin@localhost/ConservaFauna")
            
            # Criar o cursor
            cur = conn.cursor()
            
            # Executar a inserção
            cur.execute("INSERT INTO especie (nome_cientifico, nome_popular, descricao, habitat, classificacao_risco, id_ameaca) VALUES (%s, %s, %s, %s, %s, %s)", (nome_cientifico_value, nome_popular_value, descricao_value, habitat_value, classificacao_value, id_ame_value))
            # Confirmar a transação
            conn.commit()
            
            # Fechar o cursor e a conexão
            cur.close()
            conn.close()
            
            # Exibir mensagem de sucesso
            messagebox.showinfo("Inserir Registro", "Registro inserido com sucesso!")
            
            # Fechar a janela de inserção
            window_inserir_especie.destroy()
            
            # Atualizar e exibir os registros
            exibir_registros()
        
        except Exception as error:
            messagebox.showerror("Erro", f"Erro ao inserir o registro: {error}")
    
    # Criar o botão de salvar
    btn_salvar = tk.Button(window_inserir_especie, text="Salvar", command=salvar_registro)
    btn_salvar.grid(row=6, column=0, columnspan=2)

# Esconder a janela de registros inicialmente
window_registros.withdraw()

def inserir_ameaca():
    global escolha
    escolha = "Ameacas"
    # Criar a janela de inserção
    window_inserir_ameaca = tk.Toplevel(window)
    
    # Criar os campos de entrada
    lbl_descricao = tk.Label(window_inserir_ameaca, text="Descrição:")
    entry_descricao = tk.Entry(window_inserir_ameaca)
    
    lbl_nivel_ame = tk.Label(window_inserir_ameaca, text="Nível da ameaça:")
    entry_nivel_ame = tk.Entry(window_inserir_ameaca)
    
    # Posicionar os campos de entrada
    lbl_descricao.grid(row=0, column=0)
    entry_descricao.grid(row=0, column=1)
    lbl_nivel_ame.grid(row=1, column=0)
    entry_nivel_ame.grid(row=1, column=1)
    
    # Função para salvar o novo registro
    def salvar_registro():
        try:
            # Obter os valores dos campos de entrada
            descricao_value = entry_descricao.get()
            nivel_ame_value = entry_nivel_ame.get()
            
            # Conectar ao banco de dados
            conn = pg.connect("postgresql://postgres:admin@localhost/ConservaFauna")
            
            # Criar o cursor
            cur = conn.cursor()
            
            # Executar a inserção
            cur.execute("INSERT INTO ameacas (descricao, nivel_ame) VALUES (%s, %s)", (descricao_value, nivel_ame_value))
            
            # Confirmar a transação
            conn.commit()
            
            # Fechar o cursor e a conexão
            cur.close()
            conn.close()
            
            # Exibir mensagem de sucesso
            messagebox.showinfo("Inserir Registro", "Registro inserido com sucesso!")
            
            # Fechar a janela de inserção
            window_inserir_ameaca.destroy()
            
            # Atualizar e exibir os registros
            exibir_registros()
        
        except Exception as error:
            messagebox.showerror("Erro", f"Erro ao inserir o registro: {error}")
    
    # Criar o botão de salvar
    btn_salvar = tk.Button(window_inserir_ameaca, text="Salvar", command=salvar_registro)
    btn_salvar.grid(row=2, column=0, columnspan=2)

# Esconder a janela de registros inicialmente
window_registros.withdraw()

def inserir_localizacao():
    global escolha
    escolha = "Localizacao"
    # Criar a janela de inserção
    window_inserir_localizacao = tk.Toplevel(window)
    
    # Criar os campos de entrada
    lbl_id_especie = tk.Label(window_inserir_localizacao, text="ID da espécie:")
    entry_id_especie = tk.Entry(window_inserir_localizacao)
    
    lbl_latitude = tk.Label(window_inserir_localizacao, text="Latitude:")
    entry_latitude = tk.Entry(window_inserir_localizacao)
    
    lbl_longitude = tk.Label(window_inserir_localizacao, text="Longitude:")
    entry_longitude = tk.Entry(window_inserir_localizacao)
    
    lbl_descricao = tk.Label(window_inserir_localizacao, text="Descrição:")
    entry_descricao = tk.Entry(window_inserir_localizacao)
    
    # Posicionar os campos de entrada
    lbl_id_especie.grid(row=0, column=0)
    entry_id_especie.grid(row=0, column=1)
    lbl_latitude.grid(row=1, column=0)
    entry_latitude.grid(row=1, column=1)
    lbl_longitude.grid(row=2, column=0)
    entry_longitude.grid(row=2, column=1)
    lbl_descricao.grid(row=3, column=0)
    entry_descricao.grid(row=3, column=1)
    
    # Função para salvar o novo registro
    def salvar_registro():
        try:
            # Obter os valores dos campos de entrada
            id_especie_value = entry_id_especie.get()
            latitude_value = entry_latitude.get()
            longitude_value = entry_longitude.get()
            descricao_value = entry_descricao.get()
            
            # Conectar ao banco de dados
            conn = pg.connect("postgresql://postgres:admin@localhost/ConservaFauna")
            
            # Criar o cursor
            cur = conn.cursor()
            
            # Executar a inserção
            cur.execute("INSERT INTO localizacao (id_especie, latitude, longitude, descricao) VALUES (%s, %s, %s, %s)", (id_especie_value, latitude_value, longitude_value, descricao_value))
            
            # Confirmar a transação
            conn.commit()
            
            # Fechar o cursor e a conexão
            cur.close()
            conn.close()
            
            # Exibir mensagem de sucesso
            messagebox.showinfo("Inserir Registro", "Registro inserido com sucesso!")
            
            # Fechar a janela de inserção
            window_inserir_localizacao.destroy()
            
            # Atualizar e exibir os registros
            exibir_registros()
        
        except Exception as error:
            messagebox.showerror("Erro", f"Erro ao inserir o registro: {error}")
    
    # Criar o botão de salvar
    btn_salvar = tk.Button(window_inserir_localizacao, text="Salvar", command=salvar_registro)
    btn_salvar.grid(row=4, column=0, columnspan=2)


# Esconder a janela de registros inicialmente
window_registros.withdraw()

def inserir_avistamento():
    global escolha
    escolha = "Avistamentos"
    # Criar a janela de inserção
    window_inserir_avistamento = tk.Toplevel(window)
    
    # Criar os campos de entrada
    lbl_id_especie = tk.Label(window_inserir_avistamento, text="ID da Espécie:")
    entry_id_especie = tk.Entry(window_inserir_avistamento)
    
    lbl_id_localizacao = tk.Label(window_inserir_avistamento, text="ID da Localização:")
    entry_id_localizacao = tk.Entry(window_inserir_avistamento)
    
    lbl_data_avistamento = tk.Label(window_inserir_avistamento, text="Data do Avistamento:")
    entry_data_avistamento = tk.Entry(window_inserir_avistamento)
    
    lbl_observacao = tk.Label(window_inserir_avistamento, text="Observação:")
    entry_observacao = tk.Entry(window_inserir_avistamento)
    
    # Posicionar os campos de entrada
    lbl_id_especie.grid(row=0, column=0)
    entry_id_especie.grid(row=0, column=1)
    lbl_id_localizacao.grid(row=1, column=0)
    entry_id_localizacao.grid(row=1, column=1)
    lbl_data_avistamento.grid(row=2, column=0)
    entry_data_avistamento.grid(row=2, column=1)
    lbl_observacao.grid(row=3, column=0)
    entry_observacao.grid(row=3, column=1)
    
    # Função para salvar o novo registro
    def salvar_registro():
        try:
            # Obter os valores dos campos de entrada
            id_especie_value = entry_id_especie.get()
            id_localizacao_value = entry_id_localizacao.get()
            data_avistamento_value = entry_data_avistamento.get()
            observacao_value = entry_observacao.get()
            
            # Conectar ao banco de dados
            conn = pg.connect("postgresql://postgres:admin@localhost/ConservaFauna")
            
            # Criar o cursor
            cur = conn.cursor()
            
            # Executar a inserção
            cur.execute("INSERT INTO avistamentos (id_especie, id_localizacao, data_avistamento, observacao) VALUES (%s, %s, %s, %s)",
                        (id_especie_value, id_localizacao_value, data_avistamento_value, observacao_value))
            
            # Confirmar a transação
            conn.commit()
            
            # Fechar o cursor e a conexão
            cur.close()
            conn.close()
            
            # Exibir mensagem de sucesso
            messagebox.showinfo("Inserir Registro", "Registro inserido com sucesso!")
            
            # Fechar a janela de inserção
            window_inserir_avistamento.destroy()
            
            # Atualizar e exibir os registros
            exibir_registros()
        
        except Exception as error:
            messagebox.showerror("Erro", f"Erro ao inserir o registro: {error}")
    
    # Criar o botão de salvar
    btn_salvar = tk.Button(window_inserir_avistamento, text="Salvar", command=salvar_registro)
    btn_salvar.grid(row=4, column=0, columnspan=2)

# Esconder a janela de registros inicialmente
window_registros.withdraw()

def inserir_instituicao():
    global escolha
    escolha = "Instituicao"
    # Criar a janela de inserção
    window_inserir_instituicao = tk.Toplevel(window)
    
    # Criar os campos de entrada
    lbl_nome = tk.Label(window_inserir_instituicao, text="Nome:")
    entry_nome = tk.Entry(window_inserir_instituicao)
    
    lbl_sede = tk.Label(window_inserir_instituicao, text="Sede:")
    entry_sede = tk.Entry(window_inserir_instituicao)
    
    # Posicionar os campos de entrada
    lbl_nome.grid(row=0, column=0)
    entry_nome.grid(row=0, column=1)
    lbl_sede.grid(row=1, column=0)
    entry_sede.grid(row=1, column=1)
    
    # Função para salvar o novo registro
    def salvar_registro():
        try:
            # Obter os valores dos campos de entrada
            nome_value = entry_nome.get()
            sede_value = entry_sede.get()
            
            # Conectar ao banco de dados
            conn = pg.connect("postgresql://postgres:admin@localhost/ConservaFauna")
            
            # Criar o cursor
            cur = conn.cursor()
            
            # Executar a inserção
            cur.execute("INSERT INTO instituicao (nome, sede) VALUES (%s, %s)", (nome_value, sede_value))
            
            # Confirmar a transação
            conn.commit()
            
            # Fechar o cursor e a conexão
            cur.close()
            conn.close()
            
            # Exibir mensagem de sucesso
            messagebox.showinfo("Inserir Registro", "Registro inserido com sucesso!")
            
            # Fechar a janela de inserção
            window_inserir_instituicao.destroy()
            
            # Atualizar e exibir os registros
            exibir_registros()
        
        except Exception as error:
            messagebox.showerror("Erro", f"Erro ao inserir o registro: {error}")
    
    # Criar o botão de salvar
    btn_salvar = tk.Button(window_inserir_instituicao, text="Salvar", command=salvar_registro)
    btn_salvar.grid(row=2, column=0, columnspan=2)

# Esconder a janela de registros inicialmente
window_registros.withdraw()


#################################################################################################################################################


def atualizar_especie():
    global escolha
    escolha = "Especie"
    # Criar a janela de atualização
    window_atualizar_especie = tk.Toplevel(window)
    
    # Criar os campos de entrada
    lbl_id = tk.Label(window_atualizar_especie, text="ID:")
    entry_id = tk.Entry(window_atualizar_especie)
    
    lbl_nome_cientifico = tk.Label(window_atualizar_especie, text="Nome cientifico:")
    entry_nome_cientifico = tk.Entry(window_atualizar_especie)
    
    lbl_nome_popular = tk.Label(window_atualizar_especie, text="Nome popular:")
    entry_nome_popular = tk.Entry(window_atualizar_especie)
    
    lbl_descricao = tk.Label(window_atualizar_especie, text="Descrição:")
    entry_descricao = tk.Entry(window_atualizar_especie)
    
    lbl_habitat = tk.Label(window_atualizar_especie, text="Habitat:")
    entry_habitat = tk.Entry(window_atualizar_especie)
    
    lbl_classificacao = tk.Label(window_atualizar_especie, text="Classificacao de risco:")
    entry_classificacao = tk.Entry(window_atualizar_especie)
    
    lbl_id_ame = tk.Label(window_atualizar_especie, text="ID da ameaça:")
    entry_id_ame = tk.Entry(window_atualizar_especie)
    
    # Posicionar os campos de entrada
    lbl_id.grid(row=0, column=0)
    entry_id.grid(row=0, column=1)
    lbl_nome_cientifico.grid(row=1, column=0)
    entry_nome_cientifico.grid(row=1, column=1)
    lbl_nome_popular.grid(row=2, column=0)
    entry_nome_popular.grid(row=2, column=1)
    lbl_descricao.grid(row=3, column=0)
    entry_descricao.grid(row=3, column=1)
    lbl_habitat.grid(row=4, column=0)
    entry_habitat.grid(row=4, column=1)
    lbl_classificacao.grid(row=5, column=0)
    entry_classificacao.grid(row=5, column=1)
    lbl_id_ame.grid(row=6, column=0)
    entry_id_ame.grid(row=6, column=1)
    
    # Função para atualizar o registro
    def atualizar_registro():
        try:
            # Obter os valores dos campos de entrada
            id_value = entry_id.get()
            nome_cientifico_value = entry_nome_cientifico.get()
            nome_popular_value = entry_nome_popular.get()
            descricao_value = entry_descricao.get()
            habitat_value = entry_habitat.get()
            classificacao_value = entry_classificacao.get()
            id_ame_value = entry_id_ame.get()
            
            # Conectar ao banco de dados
            conn = pg.connect("postgresql://postgres:admin@localhost/ConservaFauna")
            
            # Criar o cursor
            cur = conn.cursor()
            
            # Executar a atualização
            cur.execute("UPDATE especie SET nome_cientifico = %s, nome_popular = %s, descricao = %s, habitat = %s, classificacao_risco = %s, id_ameaca = %s WHERE id = %s",
                        (nome_cientifico_value, nome_popular_value, descricao_value, habitat_value, classificacao_value, id_ame_value, id_value))
            
            # Confirmar a transação
            conn.commit()
            
            # Fechar o cursor e a conexão
            cur.close()
            conn.close()
            
            # Exibir mensagem de sucesso
            messagebox.showinfo("Atualizar Registro", "Registro atualizado com sucesso!")
            
            # Fechar a janela de atualização
            window_atualizar_especie.destroy()
            
            # Atualizar e exibir os registros
            exibir_registros()
        
        except Exception as error:
            messagebox.showerror("Erro", f"Erro ao atualizar o registro: {error}")
    
    # Criar o botão de atualizar
    btn_atualizar = tk.Button(window_atualizar_especie, text="Atualizar", command=atualizar_registro)
    btn_atualizar.grid(row=7, column=0, columnspan=2)
    
    # Preencher os campos de entrada com os valores do registro selecionado
    selecionado = ttk.selection()
    if len(selecionado) == 1:
        item = ttk.item(selecionado)
        values = item['values']
        entry_id.insert(0, values[0])
        entry_nome_cientifico.insert(0, values[1])
        entry_nome_popular.insert(0, values[2])
        entry_descricao.insert(0, values[3])
        entry_habitat.insert(0, values[4])
        entry_classificacao.insert(0, values[5])
        entry_id_ame.insert(0, values[6])
    
# Esconder a janela de registros inicialmente
window_registros.withdraw()

def atualizar_ameaca():
    global escolha
    escolha = "Ameacas"
    # Criar a janela de atualização
    window_atualizar_ameaca = tk.Toplevel(window)
    
    # Criar os campos de entrada
    lbl_id = tk.Label(window_atualizar_ameaca, text="ID:")
    entry_id = tk.Entry(window_atualizar_ameaca)
    
    lbl_descricao = tk.Label(window_atualizar_ameaca, text="Descrição:")
    entry_descricao = tk.Entry(window_atualizar_ameaca)
    
    lbl_nivel_ame = tk.Label(window_atualizar_ameaca, text="Nível da ameaça:")
    entry_nivel_ame = tk.Entry(window_atualizar_ameaca)
    
    # Posicionar os campos de entrada
    lbl_id.grid(row=0, column=0)
    entry_id.grid(row=0, column=1)
    lbl_descricao.grid(row=1, column=0)
    entry_descricao.grid(row=1, column=1)
    lbl_nivel_ame.grid(row=2, column=0)
    entry_nivel_ame.grid(row=2, column=1)
    
    # Função para atualizar o registro
    def atualizar_registro():
        try:
            # Obter os valores dos campos de entrada
            id_value = entry_id.get()
            descricao_value = entry_descricao.get()
            nivel_ame_value = entry_nivel_ame.get()
            
            # Conectar ao banco de dados
            conn = pg.connect("postgresql://postgres:admin@localhost/ConservaFauna")
            
            # Criar o cursor
            cur = conn.cursor()
            
            # Executar a atualização
            cur.execute("UPDATE ameacas SET descricao = %s, nivel_ame = %s WHERE id = %s",
                        (descricao_value, nivel_ame_value, id_value))
            
            # Confirmar a transação
            conn.commit()
            
            # Fechar o cursor e a conexão
            cur.close()
            conn.close()
            
            # Exibir mensagem de sucesso
            messagebox.showinfo("Atualizar Registro", "Registro atualizado com sucesso!")
            
            # Fechar a janela de atualização
            window_atualizar_ameaca.destroy()
            
            # Atualizar e exibir os registros
            exibir_registros()
        
        except Exception as error:
            messagebox.showerror("Erro", f"Erro ao atualizar o registro: {error}")
    
    # Criar o botão de atualizar
    btn_atualizar = tk.Button(window_atualizar_ameaca, text="Atualizar", command=atualizar_registro)
    btn_atualizar.grid(row=3, column=0, columnspan=2)
    
    # Preencher os campos de entrada com os valores do registro selecionado
    selecionado = ttk.selection()
    if len(selecionado) == 1:
        item = ttk.item(selecionado)
        values = item['values']
        entry_id.insert(0, values[0])
        entry_descricao.insert(0, values[1])
        entry_nivel_ame.insert(0, values[2])

# Esconder a janela de registros inicialmente
window_registros.withdraw()


def atualizar_localizacao():
    global escolha
    escolha = "Localizacao"
    # Criar a janela de atualização
    window_atualizar_localizacao = tk.Toplevel(window)
    
    # Criar os campos de entrada
    lbl_id = tk.Label(window_atualizar_localizacao, text="ID:")
    entry_id = tk.Entry(window_atualizar_localizacao)
    
    lbl_id_especie = tk.Label(window_atualizar_localizacao, text="ID da espécie:")
    entry_id_especie = tk.Entry(window_atualizar_localizacao)
    
    lbl_latitude = tk.Label(window_atualizar_localizacao, text="Latitude:")
    entry_latitude = tk.Entry(window_atualizar_localizacao)
    
    lbl_longitude = tk.Label(window_atualizar_localizacao, text="Longitude:")
    entry_longitude = tk.Entry(window_atualizar_localizacao)
    
    lbl_descricao = tk.Label(window_atualizar_localizacao, text="Descrição:")
    entry_descricao = tk.Entry(window_atualizar_localizacao)
    
    # Posicionar os campos de entrada
    lbl_id.grid(row=0, column=0)
    entry_id.grid(row=0, column=1)
    lbl_id_especie.grid(row=1, column=0)
    entry_id_especie.grid(row=1, column=1)
    lbl_latitude.grid(row=2, column=0)
    entry_latitude.grid(row=2, column=1)
    lbl_longitude.grid(row=3, column=0)
    entry_longitude.grid(row=3, column=1)
    lbl_descricao.grid(row=4, column=0)
    entry_descricao.grid(row=4, column=1)
    
    # Função para atualizar o registro
    def atualizar_registro():
        try:
            # Obter os valores dos campos de entrada
            id_value = entry_id.get()
            id_especie_value = entry_id_especie.get()
            latitude_value = entry_latitude.get()
            longitude_value = entry_longitude.get()
            descricao_value = entry_descricao.get()
            
            # Conectar ao banco de dados
            conn = pg.connect("postgresql://postgres:admin@localhost/ConservaFauna")
            
            # Criar o cursor
            cur = conn.cursor()
            
            # Executar a atualização
            cur.execute("UPDATE localizacao SET id_especie = %s, latitude = %s, longitude = %s, descricao = %s WHERE id = %s",
                        (id_especie_value, latitude_value, longitude_value, descricao_value, id_value))
            
            # Confirmar a transação
            conn.commit()
            
            # Fechar o cursor e a conexão
            cur.close()
            conn.close()
            
            # Exibir mensagem de sucesso
            messagebox.showinfo("Atualizar Registro", "Registro atualizado com sucesso!")
            
            # Fechar a janela de atualização
            window_atualizar_localizacao.destroy()
            
            # Atualizar e exibir os registros
            exibir_registros()
        
        except Exception as error:
            messagebox.showerror("Erro", f"Erro ao atualizar o registro: {error}")
    
    # Criar o botão de atualizar
    btn_atualizar = tk.Button(window_atualizar_localizacao, text="Atualizar", command=atualizar_registro)
    btn_atualizar.grid(row=5, column=0, columnspan=2)
    
    # Preencher os campos de entrada com os valores do registro selecionado
    selecionado = ttk.selection()
    if len(selecionado) == 1:
        item = ttk.item(selecionado)
        values = item['values']
        entry_id.insert(0, values[0])
        entry_id_especie.insert(0, values[1])
        entry_latitude.insert(0, values[2])
        entry_longitude.insert(0, values[3])
        entry_descricao.insert(0, values[4])


# Função para voltar à janela principal
def voltar():
    window_registros.withdraw()

# Esconder a janela de registros inicialmente
window_registros.withdraw()

def atualizar_avistamento():
    global escolha
    escolha = "Avistamentos"
    
    # Criar a janela de atualização
    window_atualizar_avistamento = tk.Toplevel(window)
    
    # Criar os campos de entrada
    lbl_id_avistamento = tk.Label(window_atualizar_avistamento, text="ID do Avistamento:")
    entry_id_avistamento = tk.Entry(window_atualizar_avistamento)
    
    lbl_id_especie = tk.Label(window_atualizar_avistamento, text="ID da Espécie:")
    entry_id_especie = tk.Entry(window_atualizar_avistamento)
    
    lbl_id_localizacao = tk.Label(window_atualizar_avistamento, text="ID da Localização:")
    entry_id_localizacao = tk.Entry(window_atualizar_avistamento)
    
    lbl_data_avistamento = tk.Label(window_atualizar_avistamento, text="Data do Avistamento:")
    entry_data_avistamento = tk.Entry(window_atualizar_avistamento)
    
    lbl_observacao = tk.Label(window_atualizar_avistamento, text="Observação:")
    entry_observacao = tk.Entry(window_atualizar_avistamento)
    
    # Posicionar os campos de entrada
    lbl_id_avistamento.grid(row=0, column=0)
    entry_id_avistamento.grid(row=0, column=1)
    lbl_id_especie.grid(row=1, column=0)
    entry_id_especie.grid(row=1, column=1)
    lbl_id_localizacao.grid(row=2, column=0)
    entry_id_localizacao.grid(row=2, column=1)
    lbl_data_avistamento.grid(row=3, column=0)
    entry_data_avistamento.grid(row=3, column=1)
    lbl_observacao.grid(row=4, column=0)
    entry_observacao.grid(row=4, column=1)
    
    # Função para atualizar o registro
    def atualizar_registro():
        try:
            # Obter os valores dos campos de entrada
            id_avistamento_value = entry_id_avistamento.get()
            id_especie_value = entry_id_especie.get()
            id_localizacao_value = entry_id_localizacao.get()
            data_avistamento_value = entry_data_avistamento.get()
            observacao_value = entry_observacao.get()
            
            # Conectar ao banco de dados
            conn = pg.connect("postgresql://postgres:admin@localhost/ConservaFauna")
            
            # Criar o cursor
            cur = conn.cursor()
            
            # Executar a atualização
            cur.execute("UPDATE avistamentos SET id_especie = %s, id_localizacao = %s, data_avistamento = %s, observacao = %s WHERE id = %s",
                        (id_especie_value, id_localizacao_value, data_avistamento_value, observacao_value, id_avistamento_value))
            
            # Confirmar a transação
            conn.commit()
            
            # Fechar o cursor e a conexão
            cur.close()
            conn.close()
            
            # Exibir mensagem de sucesso
            messagebox.showinfo("Atualizar Registro", "Registro atualizado com sucesso!")
            
            # Fechar a janela de atualização
            window_atualizar_avistamento.destroy()
            
            # Atualizar e exibir os registros
            exibir_registros()
        
        except Exception as error:
            messagebox.showerror("Erro", f"Erro ao atualizar o registro: {error}")
    
    # Criar o botão de atualizar
    btn_atualizar = tk.Button(window_atualizar_avistamento, text="Atualizar", command=atualizar_registro)
    btn_atualizar.grid(row=5, column=0, columnspan=2)
    
    # Preencher os campos de entrada com os valores do registro selecionado
    selecionado = ttk.focus()
    if selecionado:
        item = ttk.item(selecionado)
        values = item['values']
        entry_id_avistamento.insert(0, values[0])
        entry_id_especie.insert(0, values[1])
        entry_id_localizacao.insert(0, values[2])
        entry_data_avistamento.insert(0, values[3])
        entry_observacao.insert(0, values[4])
    else:
        messagebox.showwarning("Aviso", "Nenhum registro selecionado.")


def atualizar_instituicao():
    global escolha
    escolha = "Instituição"
    
    # Criar a janela de atualização
    window_atualizar_instituicao = tk.Toplevel(window)
    
    # Criar os campos de entrada
    lbl_id_instituicao = tk.Label(window_atualizar_instituicao, text="ID da Instituição:")
    entry_id_instituicao = tk.Entry(window_atualizar_instituicao)
    
    lbl_nome_instituicao = tk.Label(window_atualizar_instituicao, text="Nome da Instituição:")
    entry_nome_instituicao = tk.Entry(window_atualizar_instituicao)
    
    lbl_sede_instituicao = tk.Label(window_atualizar_instituicao, text="Sede da Instituição:")
    entry_sede_instituicao = tk.Entry(window_atualizar_instituicao)
    
    # Posicionar os campos de entrada
    lbl_id_instituicao.grid(row=0, column=0)
    entry_id_instituicao.grid(row=0, column=1)
    lbl_nome_instituicao.grid(row=1, column=0)
    entry_nome_instituicao.grid(row=1, column=1)
    lbl_sede_instituicao.grid(row=2, column=0)
    entry_sede_instituicao.grid(row=2, column=1)
    
    # Função para atualizar o registro
    def atualizar_registro():
        try:
            # Obter os valores dos campos de entrada
            id_instituicao_value = entry_id_instituicao.get()
            nome_instituicao_value = entry_nome_instituicao.get()
            sede_instituicao_value = entry_sede_instituicao.get()
            
            # Conectar ao banco de dados
            conn = pg.connect("postgresql://postgres:admin@localhost/ConservaFauna")
            
            # Criar o cursor
            cur = conn.cursor()
            
            # Executar a atualização
            cur.execute("UPDATE instituicao SET nome = %s, sede = %s WHERE id = %s",
                        (nome_instituicao_value, sede_instituicao_value, id_instituicao_value))
            
            # Confirmar a transação
            conn.commit()
            
            # Fechar o cursor e a conexão
            cur.close()
            conn.close()
            
            # Exibir mensagem de sucesso
            messagebox.showinfo("Atualizar Registro", "Registro atualizado com sucesso!")
            
            # Fechar a janela de atualização
            window_atualizar_instituicao.destroy()
            
            # Atualizar e exibir os registros
            exibir_registros()
        
        except Exception as error:
            messagebox.showerror("Erro", f"Erro ao atualizar o registro: {error}")
    
    # Criar o botão de atualizar
    btn_atualizar = tk.Button(window_atualizar_instituicao, text="Atualizar", command=atualizar_registro)
    btn_atualizar.grid(row=3, column=0, columnspan=2)
    
    # Preencher os campos de entrada com os valores do registro selecionado
    selecionado = ttk.focus()
    if selecionado:
        item = ttk.item(selecionado)
        values = item['values']
        entry_id_instituicao.insert(0, values[0])
        entry_nome_instituicao.insert(0, values[1])
        entry_sede_instituicao.insert(0, values[2])
    else:
        messagebox.showwarning("Aviso", "Nenhum registro selecionado.")


#################################################################################################################################################



def deletar_especie():
    global escolha
    escolha = "Especie"
    # Criar a janela de atualização
    window_deletar_especie = tk.Toplevel(window)
    
    # Criar os campos de entrada
    lbl_id = tk.Label(window_deletar_especie, text="ID:")
    entry_id = tk.Entry(window_deletar_especie)
    
    # Posicionar os campos de entrada
    lbl_id.grid(row=0, column=0)
    entry_id.grid(row=0, column=1)
   
    def atualizar_registro():
        try:
            # Obter os valores dos campos de entrada
            id_especie = entry_id.get()
            
            # Conectar ao banco de dados
            conn = pg.connect("postgresql://postgres:admin@localhost/ConservaFauna")
            
            # Criar o cursor
            cur = conn.cursor()

            cur.execute("SELECT COUNT(*) FROM avistamentos WHERE id_especie = %s", (id_especie,))
            count = cur.fetchone()[0]

            if count > 0:
                messagebox.showwarning("Aviso", "Esta espécie não pode ser excluída, pois está sendo referenciada em avistamentos.")
                return
            
            # Executar a atualização
            cur.execute("DELETE FROM especie WHERE id = %s", (id_especie,))
            
            # Confirmar a transação
            conn.commit()
            
            # Fechar o cursor e a conexão
            cur.close()
            conn.close()
            
            # Exibir mensagem de sucesso
            messagebox.showinfo("Excluir Espécie", "Espécie excluída com sucesso!")
            
            # Fechar a janela de atualização
            window_deletar_especie.destroy()
            
            # Atualizar e exibir os registros
            exibir_registros()
        
        except Exception as error:
            messagebox.showerror("Erro", f"Erro ao atualizar o registro: {error}")
    
    # Criar o botão de atualizar
    btn_atualizar = tk.Button(window_deletar_especie, text="Deletar", command=atualizar_registro)
    btn_atualizar.grid(row=3, column=0, columnspan=2)
    
    # Preencher os campos de entrada com os valores do registro selecionado
    selecionado = ttk.focus()
    if selecionado:
        item = ttk.item(selecionado)
        values = item['values']
        entry_id.insert(0, values[0])
    else:
        messagebox.showwarning("Aviso", "Nenhum registro selecionado.")

def deletar_ameaca():
    global escolha
    escolha = "Ameacas"
    # Criar a janela de exclusão
    window_deletar_ameaca = tk.Toplevel(window)
    
    # Criar os campos de entrada
    lbl_id = tk.Label(window_deletar_ameaca, text="ID:")
    entry_id = tk.Entry(window_deletar_ameaca)
    
    # Posicionar os campos de entrada
    lbl_id.grid(row=0, column=0)
    entry_id.grid(row=0, column=1)
   
    def deletar_registro():
        try:
            # Obter o valor do ID da ameaça
            id_ameaca = entry_id.get()
            
            # Conectar ao banco de dados
            conn = pg.connect("postgresql://postgres:admin@localhost/ConservaFauna")
            
            # Criar o cursor
            cur = conn.cursor()
            
            # Executar a exclusão da ameaça
            cur.execute("DELETE FROM ameacas WHERE id = %s", (id_ameaca,))
            
            # Confirmar a transação
            conn.commit()
            
            # Fechar o cursor e a conexão
            cur.close()
            conn.close()
            
            # Exibir mensagem de sucesso
            messagebox.showinfo("Excluir Ameaça", "Ameaça excluída com sucesso!")
            
            # Fechar a janela de exclusão
            window_deletar_ameaca.destroy()
            
            # Atualizar e exibir os registros
            exibir_registros()
        
        except Exception as error:
            messagebox.showerror("Erro", f"Erro ao excluir a ameaça: {error}")
    
    # Criar o botão de exclusão
    btn_excluir = tk.Button(window_deletar_ameaca, text="Excluir", command=deletar_registro)
    btn_excluir.grid(row=3, column=0, columnspan=2)
    
    # Preencher os campos de entrada com os valores do registro selecionado
    selecionado = ttk.focus()
    if selecionado:
        item = ttk.item(selecionado)
        values = item['values']
        entry_id.insert(0, values[0])
    else:
        messagebox.showwarning("Aviso", "Nenhum registro selecionado.")

    
def deletar_localizacao():
    global escolha
    escolha = "Localizacao"
    
    # Criar a janela de exclusão
    window_deletar_localizacao = tk.Toplevel(window)
    
    # Criar os campos de entrada
    lbl_id = tk.Label(window_deletar_localizacao, text="ID:")
    entry_id = tk.Entry(window_deletar_localizacao)
    
    # Posicionar os campos de entrada
    lbl_id.grid(row=0, column=0)
    entry_id.grid(row=0, column=1)
   
    def deletar_registro():
        try:
            # Obter o valor do ID da localização
            id_localizacao = entry_id.get()
            
            # Conectar ao banco de dados
            conn = pg.connect("postgresql://postgres:admin@localhost/ConservaFauna")
            
            # Criar o cursor
            cur = conn.cursor()
            
            # Executar a exclusão da localização
            cur.execute("DELETE FROM localizacao WHERE id = %s", (id_localizacao,))
            
            # Confirmar a transação
            conn.commit()
            
            # Fechar o cursor e a conexão
            cur.close()
            conn.close()
            
            # Exibir mensagem de sucesso
            messagebox.showinfo("Excluir Localização", "Localização excluída com sucesso!")
            
            # Fechar a janela de exclusão
            window_deletar_localizacao.destroy()
            
            # Atualizar e exibir os registros
            exibir_registros()
        
        except Exception as error:
            messagebox.showerror("Erro", f"Erro ao excluir a localização: {error}")
    
    # Criar o botão de exclusão
    btn_excluir = tk.Button(window_deletar_localizacao, text="Excluir", command=deletar_registro)
    btn_excluir.grid(row=3, column=0, columnspan=2)
    
    # Preencher os campos de entrada com os valores do registro selecionado
    selecionado = ttk.focus()
    if selecionado:
        item = ttk.item(selecionado)
        values = item['values']
        entry_id.insert(0, values[0])
    else:
        messagebox.showwarning("Aviso", "Nenhum registro selecionado.")

def deletar_avistamento():
    global escolha
    escolha = "Avistamentos"
    
    # Criar a janela de exclusão
    window_deletar_avistamento = tk.Toplevel(window)
    
    # Criar os campos de entrada
    lbl_id = tk.Label(window_deletar_avistamento, text="ID:")
    entry_id = tk.Entry(window_deletar_avistamento)
    
    # Posicionar os campos de entrada
    lbl_id.grid(row=0, column=0)
    entry_id.grid(row=0, column=1)
   
    def deletar_registro():
        try:
            # Obter o valor do ID do avistamento
            id_avistamento = entry_id.get()
            
            # Conectar ao banco de dados
            conn = pg.connect("postgresql://postgres:admin@localhost/ConservaFauna")
            
            # Criar o cursor
            cur = conn.cursor()
            
            # Executar a exclusão do avistamento
            cur.execute("DELETE FROM avistamentos WHERE id = %s", (id_avistamento,))
            
            # Confirmar a transação
            conn.commit()
            
            # Fechar o cursor e a conexão
            cur.close()
            conn.close()
            
            # Exibir mensagem de sucesso
            messagebox.showinfo("Excluir Avistamento", "Avistamento excluído com sucesso!")
            
            # Fechar a janela de exclusão
            window_deletar_avistamento.destroy()
            
            # Atualizar e exibir os registros
            exibir_registros()
        
        except Exception as error:
            messagebox.showerror("Erro", f"Erro ao excluir o avistamento: {error}")
    
    # Criar o botão de exclusão
    btn_excluir = tk.Button(window_deletar_avistamento, text="Excluir", command=deletar_registro)
    btn_excluir.grid(row=3, column=0, columnspan=2)
    
    # Preencher os campos de entrada com os valores do registro selecionado
    selecionado = ttk.focus()
    if selecionado:
        item = ttk.item(selecionado)
        values = item['values']
        entry_id.insert(0, values[0])
    else:
        messagebox.showwarning("Aviso", "Nenhum registro selecionado.")

def deletar_instituicao():
    global escolha
    escolha = "Instituição"
    
    # Criar a janela de exclusão
    window_deletar_instituicao = tk.Toplevel(window)
    
    # Criar os campos de entrada
    lbl_id = tk.Label(window_deletar_instituicao, text="ID:")
    entry_id = tk.Entry(window_deletar_instituicao)
    
    # Posicionar os campos de entrada
    lbl_id.grid(row=0, column=0)
    entry_id.grid(row=0, column=1)
   
    def deletar_registro():
        try:
            # Obter o valor do ID da instituição
            id_instituicao = entry_id.get()
            
            # Conectar ao banco de dados
            conn = pg.connect("postgresql://postgres:admin@localhost/ConservaFauna")
            
            # Criar o cursor
            cur = conn.cursor()
            
            # Executar a exclusão da instituição
            cur.execute("DELETE FROM instituicao WHERE id = %s", (id_instituicao,))
            
            # Confirmar a transação
            conn.commit()
            
            # Fechar o cursor e a conexão
            cur.close()
            conn.close()
            
            # Exibir mensagem de sucesso
            messagebox.showinfo("Excluir Instituição", "Instituição excluída com sucesso!")
            
            # Fechar a janela de exclusão
            window_deletar_instituicao.destroy()
            
            # Atualizar e exibir os registros
            exibir_registros()
        
        except Exception as error:
            messagebox.showerror("Erro", f"Erro ao excluir a instituição: {error}")
    
    # Criar o botão de exclusão
    btn_excluir = tk.Button(window_deletar_instituicao, text="Excluir", command=deletar_registro)
    btn_excluir.grid(row=3, column=0, columnspan=2)
    
    # Preencher os campos de entrada com os valores do registro selecionado
    selecionado = ttk.focus()
    if selecionado:
        item = ttk.item(selecionado)
        values = item['values']
        entry_id.insert(0, values[0])
    else:
        messagebox.showwarning("Aviso", "Nenhum registro selecionado.")


# Criar a barra de menu
menu_bar = tk.Menu(window)

# Definir o identificador da barra de menu
hMenu = menu_bar.winfo_id()

# Alterar o tamanho da fonte da barra de menu
ctypes.windll.user32.SendMessageW(hMenu, 0x01E5, 0, 12)  # Altere o valor '12' para ajustar o tamanho da fonte

# Criar o menu "Consulta"
menu_consulta = tk.Menu(menu_bar, tearoff=0)
# Definir a nova opção de estilo da fonte
fonte_menu = font.Font(size=10)  # Ajuste o tamanho da fonte conforme desejado

# Configurar a fonte do menu
menu_consulta.configure(font=fonte_menu)

menu_bar.add_cascade(label="Consultar", menu=menu_consulta)
menu_consulta.add_command(label="Espécie", command=consultar_especie)
menu_consulta.add_command(label="Especies em perigo", command=consultar_especies_ameacadas)
menu_consulta.add_command(label="Conserva por espécie", command=consultar_especie_conserva)
menu_consulta.add_command(label="Ameaças", command=consultar_ameacas)
menu_consulta.add_command(label="Localização", command=consultar_localizacao)
menu_consulta.add_command(label="Instituição", command=consultar_instituicao)
menu_consulta.add_command(label="Avistamentos", command=consultar_avistamento)
menu_consulta.add_command(label="Avistamentos recentes", command=consultar_avistamento_recente)
menu_consulta.add_command(label="Espécies da região amazonica", command=consultar_especies_amazonia)

#Criar o menu "Inserir"
menu_inserir = tk.Menu(menu_bar, tearoff=0)
# Definir a nova opção de estilo da fonte
fonte_menu = font.Font(size=10)  # Ajuste o tamanho da fonte conforme desejado

# Configurar a fonte do menu
menu_inserir.configure(font=fonte_menu)
menu_bar.add_cascade(label="Inserir", menu=menu_inserir)
menu_inserir.add_command(label="Espécie", command=inserir_especie)
menu_inserir.add_command(label="Ameaças", command=inserir_ameaca)
menu_inserir.add_command(label="Localização", command=inserir_localizacao)
menu_inserir.add_command(label="Avistamentos", command=inserir_avistamento)
menu_inserir.add_command(label="Instituição", command=inserir_instituicao)


#Criar o menu "Atualizar"
menu_atualizar = tk.Menu(menu_bar, tearoff=0)
# Definir a nova opção de estilo da fonte
fonte_menu = font.Font(size=10)  # Ajuste o tamanho da fonte conforme desejado

# Configurar a fonte do menu
menu_atualizar.configure(font=fonte_menu)

menu_bar.add_cascade(label="Atualizar", menu=menu_atualizar)
menu_atualizar.add_command(label="Espécie", command=atualizar_especie)
menu_atualizar.add_command(label="Ameaças", command=atualizar_ameaca)
menu_atualizar.add_command(label="Localização", command=atualizar_localizacao)
menu_atualizar.add_command(label="Avistamentos", command=atualizar_avistamento)
menu_atualizar.add_command(label="Instituição", command=atualizar_instituicao)

#Criar o menu "Deletar"
menu_deletar = tk.Menu(menu_bar, tearoff=0)
# Definir a nova opção de estilo da fonte
fonte_menu = font.Font(size=10)  # Ajuste o tamanho da fonte conforme desejado

# Configurar a fonte do menu
menu_deletar.configure(font=fonte_menu)
menu_bar.add_cascade(label="Deletar", menu=menu_deletar)
menu_deletar.add_command(label="Espécie", command=deletar_especie)
menu_deletar.add_command(label="Ameaças", command=deletar_ameaca)
menu_deletar.add_command(label="Localização", command=deletar_localizacao)
menu_deletar.add_command(label="Avistamentos", command=deletar_avistamento)
menu_deletar.add_command(label="Instituição", command=deletar_instituicao)

# Definir a barra de menu como a barra superior da janela principal
window.config(menu=menu_bar)

# Executar o loop principal do Tkinter
window.mainloop()
```
